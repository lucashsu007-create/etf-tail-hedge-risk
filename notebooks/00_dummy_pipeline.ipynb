# Dummy Pipeline

This notebook tests the project logic with simulated near-substitute ETF prices before WRDS/TAQ access is available.

```python
from src.data.load_public import make_dummy_prices
from src.features.returns import compute_simple_returns
from src.models.rolling_beta import rolling_beta
from src.features.hedge_error import hedge_error
from src.models.var_es import historical_var, historical_es

prices = make_dummy_prices(n=1000)
r_a = compute_simple_returns(prices['ETF_A'])
r_b = compute_simple_returns(prices['ETF_B'])
beta = rolling_beta(r_a, r_b, window=78)
he = hedge_error(r_a, r_b, beta)

print('VaR 5%:', historical_var(he, 0.05))
print('ES 5%:', historical_es(he, 0.05))
he.describe()
```
